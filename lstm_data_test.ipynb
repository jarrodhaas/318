```python
import pandas as pd
import numpy as np
import matplotlib as pt
import matplotlib.pyplot as plt
import time
```

```python
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
```

```python
data_train = pd.read_csv("lstm_train.csv")
data_test = pd.read_csv("rw1.csv")


seq_len = 30
data_train = data_train.as_matrix()
data_test = data_test.as_matrix()

```

```python
train = np.reshape([data_train[0:0+seq_len]], (1,seq_len))
for i in range(1, data_train.shape[0] - seq_len):
   train = np.append(train, (np.reshape([data_train[i:i+seq_len]], (1,seq_len))), axis=0)    

test = np.reshape([data_test[0:0+seq_len]], (1,seq_len))
for i in range(1, data_test.shape[0] - seq_len):
   test = np.append(test, (np.reshape([data_test[i:i+seq_len]], (1,seq_len))), axis=0)   
```

```python
result_mean = train.mean()
train -= result_mean
#test will have a slightly different mean, fix this?
test -= result_mean

```

```python
np.random.shuffle(train)
X_train = train[:, :-1]
y_train = train[:, -1]
X_test = test[:, :-1]
y_test = test[:, -1]
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
model = Sequential()
layers = [1, 30, 200, 1]

model.add(LSTM(
    layers[1],
    input_shape=(None, layers[0]),
    return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
    layers[2],
    return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(
    layers[3]))
model.add(Activation("linear"))

start = time.time()
model.compile(loss="mse", optimizer="rmsprop")
print( "Compilation Time : ", time.time() - start)
```

```python
global_start_time = time.time()
epochs = 10
```

```python
model.fit(X_train, y_train, batch_size=780, nb_epoch=epochs, validation_split=0.3, shuffle=False)
```

```python
predicted = model.predict(X_test)
predicted = np.reshape(predicted, (predicted.size,))
```

```python
fig = plt.figure(figsize=(18, 16))
ax = fig.add_subplot(111)
ax.plot(y_test[:])
plt.plot(predicted[:], 'r')
plt.show()
```
